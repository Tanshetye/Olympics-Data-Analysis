```python
df = df_cleaned.copy()
```

```python
df['Medal_Won'] = df['Medal'].notnull().astype(int)
```

```python
print(df['Medal_Won'].value_counts())
```

```python
print(df_cleaned.columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder
le_medal = LabelEncoder()
df['Medal_Won'] = le_medal.fit_transform(df['Medal']) 
```

```python
features = ['Gender', 'Year', 'Sport', 'Discipline', 'Event', 'Country']
df_model = df[features + ['Medal_Won']].dropna()
```

```python
label_encoders = {}
for col in df_model.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col])
    label_encoders[col] = le
```

```python
from sklearn.model_selection import train_test_split
X = df_model[features]
y = df_model['Medal_Won']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000, multi_class='multinomial')
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=le_medal.classes_))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
